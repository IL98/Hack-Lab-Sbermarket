**Collecting the frequency of products by each client, and also collecting average discount of each product**

```python
def get_best_product(main_dict, i, pop_products_items):
    """
    return string of a sequence of recommendations
    with a given client
    
    this is sorted by frequency of purchases of a product and by an average discount share
    
    if there are less than 50 possible products that client has ever bought, then it fills with other popular products
    """
    if main_dict.get(i) is not None:
        l = sorted(main_dict[i].items(), key=lambda item: (item[1][0], item[1][1] / item[1][0]), reverse=True)[:50]
    else:   
        l = []
    l = [x[0] for x in l]
    
    r = [x[0] for x in pop_products_items[:50-len(l)]]
    
    l.extend(r)
       
    return " ".join(map(lambda x: str(x), l))
```

collecting the product purchases for each client

```python
import glob, os
# os.chdir("/kaggle/input/test-recsys/sbermarket_tab_2_1")

dirnames = ["/kaggle/input/test-recsys/sbermarket_tab_2_1",
           "/kaggle/input/test-recsys/sbermarket_tab_2_2",
           "/kaggle/input/test-recsys/sbermarket_tab_2_3",
           "/kaggle/input/test-recsys/sbermarket_tab_2_4",
           "/kaggle/input/test-recsys/sbermarket_tab_2_5",
           "/kaggle/input/test-recsys/sbermarket_tab_2_6"]

count_user_product_dict = {}
popular_products = {}


for dirname in dirnames:
    for filename in tqdm(os.listdir(dirname)):
        products = pd.read_csv(os.path.join(dirname, filename)).loc[:,
                                            ["user_id", "order_id"]]
        for raw in products.itertuples():
            if raw.product_id != 0:
                d = count_user_product_dict.get(raw.user_id, {})
                if d.get(raw.product_id) is None:
                    d[raw.product_id] = [0, 0]     
                d[raw.product_id][0] += 1
                if raw.price != 0:
                    d[raw.product_id][1] += raw.discount / (raw.price + raw.discount)
                count_user_product_dict[raw.user_id] = d
                
                popular_products[raw.product_id] = popular_products.get(raw.product_id, 0) + 1
                
```

getting the most popular products

```python
pop_product_items = sorted(popular_products.items(), key=lambda item: item[1], reverse=True)[:50]
```

aggregating the quantities of purchases for each client

```python
dirnames = ["/kaggle/input/test-recsys/sbermarket_tab_2_1",
           "/kaggle/input/test-recsys/sbermarket_tab_2_2",
           "/kaggle/input/test-recsys/sbermarket_tab_2_3",
           "/kaggle/input/test-recsys/sbermarket_tab_2_4",
           "/kaggle/input/test-recsys/sbermarket_tab_2_5",
           "/kaggle/input/test-recsys/sbermarket_tab_2_6"]

quantity_user_product_dict = {}

for dirname in dirnames:
    for filename in tqdm(os.listdir(dirname)):
        products = pd.read_csv(os.path.join(dirname, filename)).loc[:,["user_id", "product_id", "quantity"]]
        for raw in products.itertuples():
            d = quantity_user_product_dict.get(raw.user_id, {})
            d[raw.product_id] = d.get(raw.product_id, 0) + raw.quantity
            quantity_user_product_dict[raw.user_id] = d
```

get prediction

```python
example_submission = pd.read_csv("/kaggle/input/test-recsys/sample_submission.csv")

test_user_ids = example_submission["Id"].values
```

```python
S = 0

for k in test_user_ids:
    if len(get_best_product(count_user_product_dict, k,
                           pop_product_items).split(" ")) != 50:
        S += 1
print(S)        
```

```python
submission = pd.DataFrame({
    "Id": test_user_ids,
    "Predicted": map(lambda i: get_best_product(count_user_product_dict,i,
                                                pop_product_items), test_user_ids)
})
```
